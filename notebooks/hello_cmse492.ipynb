```python
# Let's confirm that modules from the standard library can be imported and used.
# This cell loads a small CSV file and computes a basic summary.
from pathlib import Path
import csv

data_path = Path.cwd().parent / "data" / "sample_data.csv"
rows = []
with data_path.open(newline="", encoding="utf-8") as handle:
    reader = csv.DictReader(handle)
    for row in reader:
        rows.append({"name": row["name"], "value": int(row["value"])})

total = sum(item["value"] for item in rows)
average = total / len(rows) if rows else 0

print(f"Loaded {len(rows)} rows from {data_path.name}")
print(f"Total value: {total}")
print(f"Average value: {average:.2f}")
```
